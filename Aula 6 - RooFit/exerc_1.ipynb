```c++
// Criando par√¢metros da PDF crystall ball
RooRealVar x("x", "x", 0, -10, 10);
RooRealVar x0("x0", "x0", 1, -10, 10);
RooRealVar sigma("sigma", "sigma", 1, -10, 10);
RooRealVar alpha("alpha", "alpha", 1, -10, 10);
RooRealVar n("n", "n", 1, -10, 10);

// Criando a PDF
RooCBShape cb("cb", "Crystall Ball", x, x0, sigma, alpha, n);
```

```c++
//Plotando a PDF
RooPlot* frame = x.frame();
cb.plotOn(frame);
frame->Draw();
c1->Draw();
```

```c++
// Gerando eventos com toy MC
RooDataSet* toyData = cb.generate(x, 10000);
```

```c++
// Plotando os dados com a pdf
RooPlot* frame = x.frame();
toyData->plotOn(frame);
cb.plotOn(frame);
frame->Draw();
c1->Draw();
```

```c++
// Fazendo o fit
cb.fitTo(*toyData);
```

```c++
// Extra: fazendo o mesmo para a PDF exponencial

//Gerando Param√™tros
RooRealVar xExp("xExp", "xExp", 2, -10, 10);
RooRealVar cExp("cExp", "cExp", 2, -10, 10);

//Gerando a PDF
RooExponential exp("exp", "exponential", xExp, cExp);

//Gerando dados do toy MC
RooDataSet* toyDataExp = exp.generate(xExp, 10000);

//Plotando
RooPlot* frame = xExp.frame();
toyDataExp->plotOn(frame);
exp.plotOn(frame);
frame->Draw();
c1->Draw();
// Fazendo o fit
exp.fitTo(*toyDataExp);
```
