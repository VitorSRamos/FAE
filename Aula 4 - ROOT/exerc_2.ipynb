```c++
// instanciando canvas e gráfico usando dados do arquivo
TCanvas canvas_3;
TGraph gd("data/graphdata.txt");


// definindo parâmetros
gd.SetMarkerStyle(61); // Formato do marcador, com linewidth um pouco maior
gd.SetMarkerSize(2.0); // Tamanho do marcador 
gd.SetMarkerColor(3); // Cor do marcador

gd.Draw("AP"); // "A" especifica que devem ser denhados os eixos, e "P" especifíca que devem ser denhados os pontos

canvas_3.Draw();
```
