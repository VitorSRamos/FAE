```c++
// instanciando função e canvas
TCanvas canvas_1("c1", "c1");
TF1 f1("f1", "sin(x)/x", 0, 10);

//plot
f1.Draw();
canvas_1.Draw();
```

```c++
// Instanciando função com parâmetros e novo canvas
TCanvas canvas_2("c", "c");
TF1 f2("f2", "[p0]*sin([p1]*x)/x", 0., 10.);
```

```c++
// definindo um conjunto de parâmetros e plotando
f2.SetParameter(1,2);
f2.Draw();
canvas_2.Draw();
```

```c++
//definindo um segundo conjunto de parâmetros e plotando
f2.SetParameter(2,3);
f2.Draw();
canvas_2.Draw();
```

```c++
// Mudando cor para azul e plotando
f2.SetLineColor(4);

f2.Draw();
canvas_2.Draw();
```
