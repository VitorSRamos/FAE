```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_table("/content/drive/MyDrive/UERJ/IADFAETA/Aula 8 - Monte Carlo/PYTHIA/mu_data.dat", delimiter='\s+', names=['mu_1',  'mu_2']) 

display(df)
```

```python
# Plotando o histograma
fig, ax = plt.subplots(figsize = (10,6))

plt.hist(df['mu_1'], 100) 

#detalhes do plot
ax.set_title(r'Múon 1')
ax.set_ylabel('Contagem de eventos')
ax.set_xlabel('Massa Invariante')
plt.grid()
```

```python
# Plotando o histograma
fig, ax = plt.subplots(figsize = (10,6))

plt.hist(df['mu_2'], 100) 

#detalhes do plot
ax.set_title(r'Múon 2')
ax.set_ylabel('Contagem de eventos')
ax.set_xlabel('Massa Invariante')
plt.grid()
```
